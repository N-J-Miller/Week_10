# 1. Write a function that calculates the dot product manually. 
It should accept two vectors and perform that calculation. Be sure to use try except and error handling for basic issues that might occur.


```python
# Dot Product Formula: dot product of a and b = ax*bx + ay*by
# or dot product of a and b = ||a||*||b||cosa
# For a 2D vector, this is the same as matrix multiplication to get a scalar value.

def dot_product (a,b):
    try:
        if len(a)<len(b):
            r = len(a)
        else: 
            r = len(b)
                
        dot = 0 
        for i in range(r):
            dot = dot + a[i]*b[i]
        print(dot)
    except Exception as e:
            print(e)

        
v = np.array([1,2,3])
v2 = np.array([7,8,9])


dot_product(v,v2)
```

# 2. In a markdown section in your notebook, explain what one hot encoding is and why we use it. 
Then, manually calculate a one-hot encoding for the Belize_Airbnb_listings.csv neighborhood_group column and merge it back into the dataframe. Note: manually means donâ€™t use scikit learn or keras or get_dummies

```python
# make a list of neighborhoods in 'neighbourhood_group' column

neighborhoods = belize_df['neighbourhood_group'].unique()
print(neighborhoods)
neighborhoods_series = belize_df['neighbourhood_group']

# take each element in group column and assign it to an integer
neighborhoods_to_int = dict((g, i) for i, g in enumerate(neighborhoods))
int_to_neighborhoods = dict((i, g) for i, g in enumerate(neighborhoods))

# encode with integers
integer_encoded = [neighborhoods_to_int[data] for data in neighborhoods_series]

onehot_encoded = list()
for value in integer_encoded:
    
    #create a matrix for each neighborhood group (6)
    neighborhood = [0 for _ in range(len(neighborhoods))]
    neighborhood[value] = 1
    onehot_encoded.append(neighborhood)
    
# Now we have a list. We turn it into a series for better use with Pandas.
encoded_series = pd.Series(onehot_encoded)

# we can now take our series and concat it with the data frame
onehot_belize = pd.concat([belize_df, encoded_series], axis="columns")
onehot_belize.columns = ['id','name','host_id','host_name','neighbourhood_group','neighbourhood','latitude','longitude','room_type','price','minimum_nights','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count','availability_365','Neighbourhood']
onehot_belize.head()
```
